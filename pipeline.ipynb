```python
import os
import sys

def get_environment():
    # Проверка на Google Colab
    if 'COLAB_GPU' in os.environ:
        return "Google Colab"
    
    # Проверка на Kaggle
    elif 'KAGGLE_KERNEL_RUN_ID' in os.environ:
        return "Kaggle"
    
    # Проверка на локальной машине
    elif 'HOME' in os.environ or 'USERPROFILE' in os.environ:
        return "Local PC"
    
    # Дополнительная проверка для WSL
    elif 'WSL_DISTRO_NAME' in os.environ:
        return "WSL"
    
    # Если окружение не найдено
    else:
        return "Unknown Environment"


print(f"Running on: {get_environment()}")
```
